# Apply

```python
import pandas as pd
import numpy as np
```

```python
exam_data={'name':['Anastasia','Dima','Katherine','James','Emily','Michael','Matthew','Laura','Kevin','Jonas'],'score':[12.5,9,16.5,np.nan,9,20,14.5,np.nan,8,19],
           'attempts':[1,3,2,3,2,3,1,1,2,1],'fee_given':[750,650,200,525,0,100,500,300,250,150]}

lebels=list('abcdefghij') 

df=pd.DataFrame(exam_data,index=lebels)
df
```

```python
#apply
#if score greater than 7 ,then yes or no

df['qualify']=df['score'].apply(lambda x:'YES' if x>=7 else 'NO')

df
```

```python
#actual fee is 500, add a column that will show remaining amount

def remaining(x):
    return x-500

df['balance']=df['fee_given'].apply(lambda x:remaining(x))
df
```

```python
#if balance is +ve then one column will show get/pay

df['action']=df['balance'].apply(lambda x:'GET' if x>=0  else 'PAY')
df
```

```python
#or

def action(x):
    if x>0:
        return 'GET'
    elif x<0:
        return 'PAY'
    else:
        return 'NO ACTION'

df['action_1']=df['balance'].apply(lambda x:action(x))  
df
```

```python
#section A,B,C if score greater than 10 then A,between 5-9 then B,else C

def section_selection(x):
    if x>10:
        return 'A'
    elif x>=9 and x<=10:
        return 'B'
    else:
        return 'C'

df['section']=df['score'].apply(lambda x:section_selection(x))
df
```

```python
df.drop('SECTION',axis=1,inplace=True)
```

```python
#first row
df.apply(lambda x:x[0])
```

```python
#first column
df.apply(lambda x:x[0],axis=1)
```

```python
import pandas as pd

sample_data={'X':[78,85,96,80,86],'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}

df=pd.DataFrame(sample_data)
print(df)
def greaterbetweenXnY(row):
    x=row['X']
    y=row['Y']
    if x>y:
        return x
    else:
        return y

df1=df.apply(lambda x:greaterbetweenXnY(x),axis=1)

print(df1.values)

kwargs={'greater':df1.values}

df=df.assign(**kwargs)
df
```

```python
import pandas as pd

sample_data={'X':[78,85,96,80,86],'Y':[84,94,89,83,86],'Z':['86','97','96','72','83']}

df=pd.DataFrame(sample_data)

def greaterbetweenXnY(row):
    x=row['X']
    y=row['Y']
    if x>y:
        return x
    else:
        return y
    

df['greater_new']=df[['X','Y']].apply(lambda x:greaterbetweenXnY(x),axis=1)

df['Z']=df['Z'].apply(pd.to_numeric)

df['x+z']=df['X']+df['Z']
df
```
